## Comandos customizables

Para generar un comando con sh utilizamos simplemente:

- comando = sh.Command(< comando>)

Y para usarlo, usamos como si fuera una funcion con sus argumentos (no nombrados):

- comando(*args)

El mencionado es el uso mas basico. Sin embargo existen un conjunto de parametros especiales que se pueden utilizar con los comandos, y estos son parametros nombrados que empiezan con underscore. Aqui hay una tabla de estos parametros:

Son de particular interes _err, _out, _cwd , dado que estos permiten mostrar el error en pantalla, la salida del algoritmo en pantalla y la ejecucion del algoritmo desde el directorio de interes.

```python
import sh
def spherical_popcorn_void_finder(
    *, bin_path, conf_file_path, work_dir_path, cores
):
    """
    Runs the Popcorn-Spherical Void Finder using MPI.

    Parameters
    ----------
    bin_path : pathlib.Path
        Path to the folder containing the binary file for the void finder.
    conf_file_path : pathlib.Path
        Path to the configuration file used to run the void finder.
    work_dir_path : str
        Path to the working directory.
    cores : int
        Number of cores to run the mpi run.

    Returns
    -------
    output : str
        Standard output from the run.

    Notes
    -----
    The mpi command is the following: mpirun -np <cores> --bind-to core
    """
    # Reference to mpi
    # https://docs.oracle.com/cd/E19356-01/820-3176-10/ExecutingPrograms.html


    # Command will be executed from work_dir_path path.

    mpirun = sh.Command("mpirun")
    svf_popcorn = sh.Command(bin_path / "svf")

    if type(cores) is int:
        mpirun(
            "-np", str(cores), "--bind-to", "core",
            svf_popcorn(
                f"config={conf_file_path}",
                _out=lambda line: print(line, end=""), #In sh this distinguish keywords from actual args
                _err_to_out=True
                ),
            _cwd=work_dir_path,
    
        )
    if type(cores) is None:
        print("Running without MPI")

        svf_popcorn(
            f"config={conf_file_path}",
            _out=lambda line: print(line, end=""), #In sh this distinguish keywords from actual args
            _err_to_out=True
            )
```

Concentrarse en esta funcion:

Aca hay 2 comandos:

- mpirun : corre mpi
- svf_popcorn : corre el spherical void finder

Como podemos ver los comandos estan anidados: mpirun( svf_popcorn). Como svf_popcorn tiene los outputs es que este comando es el que tiene las salidas asociadas al mismo.

Sin embargo, aunque este comando hace funcionar a svf, no hace que funcione correctamente MPI , de hecho hace un uso por defecto del mismo, pero a la larga trae errores.

La explicacion es esta:

In sh, when you call a command like popcorn(...) inside another command (mpirun(...)), it tries to execute popcorn immediately and pass its output as an argument to mpirun. Instead, mpirun needs popcorn as an executable, not its output.

mpirun(
    "-np", str(cores), "--bind-to", "core",
    popcorn( ... )  # ‚ùå This executes popcorn first, instead of passing it as a command
)

You should pass popcorn and its arguments as separate command-line arguments, instead of executing popcorn first.
