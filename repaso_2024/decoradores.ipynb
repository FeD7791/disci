## Decorador

Un decorador es una funcion d que recibe como parametro otra funcion a y retorna una nueva funcion r
- d: funcion decoradora.
- a: funcion a decorar.
- r: funcion decorada.

- La sintaxis tipica de un decorador es: r = d(a)
