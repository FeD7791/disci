## Poetry book
Para crear nuestro primer paquete lo que vamos a hacer primero es un projecto simple. Este es un contador de palabras y cuenta utilizando la funcion count la cantidad de palabras que aparecen en una lista. Porsupuesto lo que vamos a tener que hacer primero es tomar un texto, convertirlo en lista y luego utilizar count. Finalmente lo empaquetaremos.

Primero, empezamos guardando en un archivo de texto el zen de python. El mismo puede obtenerse escribiendo: import this
Concretamente hacemos lo siguiente para guardar este texto en un archivo de texto:

py -c "import this" > zen.txt

Aqui, -c nos permite pasar un string para que python lo ejecute.
Mientras el signo > redirige la salida del output ejecutado a un archivo, muy util!

# Empaquetado utilizando setuptools

Aca:
- pyproject-build ---> front end : A build frontend is a tool you run to initiate building a package
from source code. The build frontend provides a user interface and integrates
with the build backend via the hook interface.

- setuptools ---> back end :A build backend is a Python object that provides several required
and optional hooks that implement packaging behavior. The core build backend
interface is defined in PEP 517 (http://mng.bz/o5Rj).

1. Hay que tenes instalado build, esto se hace con pip : pip install build. Esto nos habilita a utilizar el comando:  pyproject-build
Este comando se utiliza para crear el proyecto. Si corremos este comando obtendremos error porque nos especifica primero tener un setup.py o un pyproject.toml
2. Creamos para este caso un pyproject.toml . Este es un nombre estandar del archivo toml.
3. Corremos el comando pyproject-build . Nota, obtendremos muchos archivos con nombre UNKNOWN
4. En pyproject.toml agregamos los siguiente:
        <div style='color:yellow'>[build-system] </div>---> Describe el build system
        <div style='color:yellow'>requires = ["setuptools", "wheel"]</div> ---> Lista de dependencias, anotadas como strings que deben instalarse. 
        <div style='color:yellow'>build-backend = "setuptools.build_meta" </div> ---> Backend
En cuanto a las dependencias tener en cuenta que una build de setup tools requiere tener instaladas tanto setuptools y wheel
5. Volver a correr el comando: **pyproject-build**
6. Para agregar la metadata del autor empezamos crando en el directorio raiz el archivo de nombre **setup.cfg**, *For the Setuptools backend, you can specify static metadata in an INI-style file called setup.cfg in the root directory of your project*
7. Agregamos el nombre y la version en el archivo **setup.cfg**
<div style='color:cyan'>[metadata]</div>
<div style='color:cyan'>name = first-python-package</div>
<div style='color:cyan'>version = 0.0.1</div>
<div style='color:cyan'>author = Federico Bustillos</div>
<div style='color:cyan'>author_email = jorgefedericobustillosrava@gmail.com</div>
Volvemos a correr el comando *pyproject-build*

Otras cosas a agregar son:
- description
- long_description 
- long_description_content_type

8. Removemos el directorio **dist** y volvemos a correr el comando *pyproject-build* . Esto nos genera nuevamente el directorio dist con los elementos .whl y tar.gz con los nombres adecuados. Para verificar el contenido de los archivos tar podemos escribir: *tar -xzf first-python-package-0.0.1.tar.gz* con el nombre del arhivo correspondiente porsupuesto.

9. Creamos el README.md y agregamos elementos en el
10. En el archivo *setup.cfg* agregamos los elementos correspondientes de:
- long_description ---> Aqui utilizamos la directiva especial **file:** . Esta directiva acepta un path a un archivo, en este caso el README.md relativo a setup.cfg
- long_description_content_type ---> En esta seccion se indica el tipo del archivo especificado en path

Ejecutamos nuevamente el comando *pyproject-build* luego inspeccionamos PKG-INFO y podremos ver la presencia de estos elementos agregados:
- En Summary tenemos la description
- El tipo del archivo esta especificado en Description-Content-Type
- Los contenidos del README.md se muestran al final del archivo

*When you update your README file, those changes will be pulled into the next version of the package you build. This automation reduces the issue of remembering to update your documentation in multiple places.*

11. Agregamos la licencia, para ello tenemos que:
- Crear un archivo LICENSE.txt 
- En el archivo setup.cfg agregamos:
  - license = MIT
  - license_files = LICENSE 
  - classifiers = License :: OSI Approved :: MIT License (ver https://pypi.org/classifiers/)

## Source code and discovery

Setup tools necesita conocer lo siguiente:

- En que directorios buscar por paquetes
- Los nombres de los sup-paquetes y una directiva para encontrarlos de forma recursiva
- Como mapear distintos paquetes encontrados a nombres distintos de importacion, si se necesita

Para ello se utiliza:

<div style='color:green'>[options]</div>--->*This section provides additional options for Setuptools package
builds.*
<div style='color:green'>[options].package_dir</div>---> *This is a list of key-value pairs to map discovered directories to import paths. An empty key means the “root,” such that any directory mapped to the root will be removed from the import path and only its child directories will be included*
<div style='color:green'>[options].packages</div>--->This is either an explicit list of packages or the special
find: directive that tells Setuptools to recursively search for any packages.
find: is often the best choice, because you won’t need to update it if you add
new packages later.
<div style='color:green'>[options.packages.find]</div>---> *This section provides options to the Setuptools
package discovery process triggered by the find: directive*
<div style='color:green'>[options.packages.find].where</div>--->This tells Setuptools which directory to look
in for packages.
<div style='color:green'>[options.packages.find].exclude</div>--->This will exclude any (sub)packages that begin with test from the distribution
package.

Why do I see the setup.py file used so widely?
The setup.py file has been a part of the Python packaging ecosystem for ages—it
even predates Setuptools. Introduced in PEP 229 (https://www.python.org/dev/peps/
pep-0229/) in 2000, its goal was to centralize where packaging configuration happens.
You’ll likely see that a wide variety of packages still use it, and although it’s
still necessary for some use cases, the new build workflow and tools you learned in
chapters 2 and 3 are intended to replace setup.py in the long term. For pure-Python
packages that don’t need to determine any dynamic information at build time, you
can use pyproject.toml to define your build and setup.cfg for configuration when using
Setuptools as your build backend.


En el libro se da un ejemplo concreto de un codigo creado en cyton y que hay que utilizar un ejecutable.
Para correr Cython es indispensable contar con el archivo setup.py pero este basta con contener algunas lineas nada mas.
En este <a href='https://github.com/daneah/publishing-python-packages/blob/main/ch04/first-python-package/setup.py'>Link</a> tenemos el repositorio del libro, si nos fijamos el ch4 tiene el archivo setup.py

```python
import os
import pathlib

import ez_setup

ez_setup.use_setuptools()

from setuptools import setup  #Aca utiliza setuptools

with open("README.md") as fp:
    LONG_DESCRIPTION = fp.read()

REQUIREMENTS = ["numpy", "pandas", "scipy", "attrs", "matplotlib"]

setup(
    name="Pyedra",
    version=VERSION,
    description="Implementation of phase function for asteroids in Python",
    long_description=LONG_DESCRIPTION,
    long_description_content_type="text/markdown",
    author="Milagros Colazo",
    author_email="milirita.colazovinovo@gmail.com",
    url="https://github.com/milicolazo/Pyedra",
    py_modules=["ez_setup"],
    packages=[
        "pyedra",
        "pyedra.datasets",
    ],
    license="The MIT License",
    install_requires=REQUIREMENTS,
    keywords=["pyedra", "asteroid", "phase function"],
    classifiers=[
        "Development Status :: 4 - Beta",
        "Intended Audience :: Education",
        "Intended Audience :: Science/Research",
        "License :: OSI Approved :: MIT License",
        "Operating System :: OS Independent",
        "Programming Language :: Python",
        "Programming Language :: Python :: 3.8",
        "Programming Language :: Python :: Implementation :: CPython",
        "Topic :: Scientific/Engineering",
    ],
    include_package_data=True,
)
```

fijarse que tiene una seccion de setup que es muy similar a lo que tenemos en setup.cfg
