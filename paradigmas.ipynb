# Paradigmas

Source: Wikipedia

Existen muchos paradigmas sobre los cuales se basan los lenguajes de programacion, estos se dividen en un arbol. 

Principalmente los paradigmas son 2:

1. Imperativo : el programador instruye a la máquina cómo cambiar su estado

    - Procedimental : que agrupa las instrucciones en procedimientos
    - Orientado a objetos : agrupa las instrucciones con la parte del estado en el que operan

2. Declarativo : el programador simplemente declara las propiedades del resultado deseado, pero no cómo calcularlo

    - Logico : el resultado deseado se declara como el valor de una serie de aplicaciones de función
    - Funcional : el resultado deseado se declara como la respuesta a una pregunta sobre un sistema de hechos y reglas
    - Matematico : el resultado deseado se declara como la solución de un problema de optimización
    - Reactivo : se declara el resultado deseado con flujos de datos y la propagación del cambio

Hoy en dia muchos de los lenguajes se desarrollan en una combinacion de estos paradigmas.

Python corresponde concretamente al paradigma de Orientado a Objetos.

### El concepto de Shared state

*A **shared state** is any variable, object, or memory space that exists in a shared scope. Any non-constant variable used by multiple separate scopes, including the global scope and closure scopes, is considered to be in a shared state. In functional programming, shared states should be avoided.*

En el paradigma de lenguaje orientado a objetos no se comparte estado (No share)
En programacion funcional no hay estado (No State)

### Concepto de Acoplamiento y cohesion

*La **cohesión** mide el grado de conexión funcional entre los constituyentes o elementos de un mismo módulo. El **acoplamiento** es una medida del grado de interdependencia entre los módulos de un sistema. Lo deseable es tener módulos con poco acoplamiento (o independiente entre sí)*

## El paradigma orientado a objetos

Source: https://es.wikipedia.org/wiki/Programaci%C3%B3n_orientada_a_objetos
Source 2: http://cidecame.uaeh.edu.mx/lcc/mapa/PROYECTO/libro10/111_clases_y_objetos.html#:~:text=Un%20objeto%20es%20una%20entidad,un%20objeto%20debe%20ser%20claro.

Este es un paradigma que parte del concepto de objetos como base. En este paradigma los datos relacionados con algun elemento se guardan en un componente que representa un estado. Se puede utilizar una especie de logica para modificar el estado del componente. Los estados y logica se guardan en componentes de memoria.

Los objetos son entidades que tienen un determinado *estado*, *comportamiento* e *identidad*.
    

- Identidad : es una propiedad de un objeto que lo diferencia del resto; En el mundo real, dos personas pueden compartir las mismas características: nombre, fecha de nacimiento, ocupación, sin embargo, no hay duda de que son individuos distintos, con su propia identidad. El mismo concepto es válido para los objetos. Aunque dos objetos compartan el mismo estado (atributos y relaciones), siguen siendo objetos independientes con su propia identidad.

- Estado : El estado de un objeto es implantado por un conjunto de propiedades denominado *atributos*, junto con los valores de estas propiedades y los vinculos que posee el objeto con otros.

- Comportamiento : La conducta determina cómo un objeto actúa y reacciona a las solicitudes de otros. La conducta de un objeto es representada por las operaciones que un objeto realiza. La conducta esta relacionada a los *metodos* de una clase.


### Diferencias entre programacion estructurada y OOP

La programación estructurada prima el concepto de procedimientos o funciones sobre el de estructuras (se emplean principalmente funciones que procesan datos). 

La programación orientada a objetos, en cambio, primero se definen los objetos o estructuras para posteriormente solicitar la ejecución de sus métodos.

### Notas

1. Clase: Es una plantilla para crear otros objetos, en ella estan definidos los atributos y metodos del objeto.
    - Python implementa objetos con clases
    - Las clases instancian (crean) objetos
    - En Python las clases son objetos, de hecho , todo es un Objeto en Python
2. Herencia: Permite a otra clase heredar atributos y metodos de otra clase, si B hereda de A:
    - B puede utilizar los atributos y metodos publicos de A
    - Los metodos y atributos privados de A no son visibles pero B puede interactuar con ellos por medio de metodos publicos heredados de A
3. Estado: Es una variable que se declara privada, que puede ser únicamente accedida y alterada por un método del objeto, y que se utiliza para indicar distintas situaciones posibles para el objeto (o clase de objetos). No es visible al programador que maneja una instancia de la clase.
4. Mensaje: Una comunicación dirigida a un objeto, que le ordena que ejecute uno de sus métodos con ciertos parámetros asociados al evento que lo generó.
    - Cuando un objeto no sabe que hacer se emite un mensaje al objeto colaborador
    - El colaborador responde con otro mensaje
    - Los mensajes y las respuestas son objetos
    - Todo objeto en Python se comunica con otro en el siguiente esquema: objeto.mensaje(Colaborador)
