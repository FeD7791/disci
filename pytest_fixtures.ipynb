# Useful pytest fixtures



<table border="1" cellspacing="0" cellpadding="6">
  <thead>
    <tr>
      <th>Fixture</th>
      <th>Description</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td><code>tmp_path</code></td>
      <td>Provides a temporary directory as a <code>pathlib.Path</code> object. Automatically cleaned up after the test.</td>
    </tr>
    <tr>
      <td><code>tmp_path_factory</code></td>
      <td>Factory for creating multiple temporary directories in a test session. Useful for setup in <code>session</code>-scoped fixtures.</td>
    </tr>
    <tr>
      <td><code>tmpdir</code></td>
      <td>Legacy version of <code>tmp_path</code>, returns <code>py.path.local</code> objects. Still supported but <code>tmp_path</code> is preferred.</td>
    </tr>
    <tr>
      <td><code>tmpdir_factory</code></td>
      <td>Legacy version of <code>tmp_path_factory</code>. Provides a factory for creating multiple <code>tmpdir</code> objects.</td>
    </tr>
    <tr>
      <td><code>capsys</code></td>
      <td>Captures output to <code>sys.stdout</code> and <code>sys.stderr</code> during a test function. Useful for testing CLI output.</td>
    </tr>
    <tr>
      <td><code>capfd</code></td>
      <td>Captures output written to file descriptors 1 and 2 (stdout/stderr). Useful for lower-level output testing.</td>
    </tr>
    <tr>
      <td><code>caplog</code></td>
      <td>Captures logging output for assertions on logged messages.</td>
    </tr>
    <tr>
      <td><code>monkeypatch</code></td>
      <td>Temporarily modifies attributes, dictionaries, environment variables, or even imports for the duration of the test.</td>
    </tr>
    <tr>
      <td><code>request</code></td>
      <td>Provides access to the test context, including the function, class, module, or session scope. Often used inside custom fixtures.</td>
    </tr>
    <tr>
      <td><code>cache</code></td>
      <td>Provides access to a caching mechanism for sharing data across test runs (e.g., to avoid expensive setup repeatedly).</td>
    </tr>
    <tr>
      <td><code>recwarn</code></td>
      <td>Captures and asserts warnings triggered during a test.</td>
    </tr>
    <tr>
      <td><code>pytestconfig</code></td>
      <td>Gives access to the pytest configuration object, useful for checking CLI options or config values.</td>
    </tr>
    <tr>
      <td><code>record_property</code></td>
      <td>Records a property value for reporting plugins (e.g., JUnit XML).</td>
    </tr>
    <tr>
      <td><code>record_xml_attribute</code></td>
      <td>Records additional XML attributes for reporting (e.g., for CI pipelines).</td>
    </tr>
    <tr>
      <td><code>record_testsuite_property</code></td>
      <td>Records key-value pairs at the testsuite level for XML output.</td>
    </tr>
    <tr>
      <td><code>doctest_namespace</code></td>
      <td>Lets you inject values into the namespace where doctests are executed.</td>
    </tr>
  </tbody>
</table>
